# EDA on Titanic Dataset

```python
#importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Reading the data
df=pd.read_csv('F:\\Assignment\\Csv\\titanic_data1.csv')
print(df.shape)
df.head()
```

```python
df.info()# to see the high level overview of datatypes
```

```python
df.isnull().mean()*100 # check for null values
```

```python
# As we can see that here in cabin column 77% values missing.So we drop this column
#Also we drop the column passenger id and name columns
df=df.drop(['PassengerId','Name','Cabin'],axis=1)
df.shape
```

```python
# We fill the values age column by median
df=df.fillna(df['Age'].median())
df.isnull().sum()
```

```python

sns.countplot(df['Survived'],hue='Sex',data=df)
```

```python
print(df['Survived'].sum()/(df['Survived'].count())) # % of survived people
df.groupby(['Survived','Sex'])['Survived'].count() # We can see that more female  survived
```

```python
sns.countplot(df['Pclass'],hue='Survived',data=df)
```

```python
sns.countplot(df['SibSp'],hue='Survived',data=df)
```

```python
sns.countplot(df['Parch'],hue='Survived',data=df)
```

```python
sns.boxplot(data=df) # We can observe that Fare column contains more outliers
```

```python
sns.distplot(df['Age']) # more people of age between 10-60 years
```

```python
sns.distplot(df['Fare']) # We can see that fare is left skewed
```

```python
sns.pairplot(df,hue="Survived") # Pairplot to fond the relationship between variables
```

```python
sns.heatmap(df.corr(),annot=True) # To see the how variables are corelated to each others
```

# EDA on BM Dataset

```python
#Reading the data
df=pd.read_csv("F:\Assignment\Csv\BM_data.csv")
print(df.shape)
df.head()
```

```python
df.info()# to see the high level overview of datatypes
```

```python
df.isnull().mean()*100 # check for null values
```

```python
# filling nan values of Item_Weight by mean and Outlet_size by mode
df['Outlet_Size'].unique()
df['Item_Weight']=df['Item_Weight'].fillna(df['Item_Weight'].mean())
df['Outlet_Size']=df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0])
df.isnull().sum()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,3,1)
sns.countplot(df['Item_Fat_Content']) # there are more items having low fat content
plt.subplot(1,3,2)
sns.countplot(df["Outlet_Size"]) #Medium Size Outlet is more
plt.subplot(1,3,3)
sns.countplot(df['Outlet_Location_Type']) # Tier3 location is more
```

```python
plt.figure(figsize=(20,7))
sns.countplot(df['Item_Type']) 
df['Item_Type'].unique()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(df['Outlet_Type']) #Supermarket type1 is more
```
